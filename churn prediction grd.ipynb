```python
# ENCODING OF DATA BEFORE THE ML MODELLING
```

```python
df["Churn"] = df["Churn"].map({"Yes":1,"No":0})
```

```python
df["Churn"].value_counts()
```

```python
df = pd.get_dummies(df, drop_first=True)
# Converts categorical columns into multiple binary columns
df = df.astype(int)
df.head()
```

```python
# # MODEL BUILDING
# I'M Training a model to predict: Will customer churn (1) or not (0)?
```

```python
x= df.drop("Churn",axis=1)#Input (X) → data used to predict
y=df["Churn"]#Output (y) → what we want to predict
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

from sklearn.linear_model import LogisticRegression 
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred[:10]
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:",accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
# 914>>Correctly Predicted "NO"
# 119>>Wrongly Predicted "False Alarms"
# 181>> Missed Churn Customers
# 193>> Correctly Caught churn
```

```python
# Although Logistic Regression achieved reasonable accuracy,
# it struggled to correctly identify churn customers (low recall).
# To better capture complex patterns and improve churn detection,
# we use a Decision Tree model.
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train, y_train)
```

```python
y_pred_dt = dt_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Decision Tree Accuracy:", accuracy_dt)

cm_dt = confusion_matrix(y_test, y_pred_dt)
print(cm_dt)
```

```python
# # IMP Churn Detection :
# Logistic:
# Caught churn → 193
# Missed churn → 181

# Decision Tree:
# Caught churn → 196
# Missed churn → 178
# False Alarms

# Logistic:
# False positives → 119
# Decision Tree:
# False positives → 200

# Final Interpretation
#  Logistic Regression:
# Better accuracy
# Fewer false alarms

#  Decision Tree:
# Slightly better churn detection
# But more false alarms

```
